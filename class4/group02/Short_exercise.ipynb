```
##create data
# color -- blue, dark blue, light blue, others
# sleeve short /long
# women/men
# pure color/not pure
# texture hard, medium hard, medium soft, soft
pic1 = [0, 0, 1, 1, 1]
pic2 = [1, 1, 1, 0, 0]
pic3 = [1, 0, 1, 1, 1]
pic4 = [2, 0, 0, 0, 2]
pic5 = [3, 1, 0, 1, 3] 
pics = [pic1, pic2, pic3, pic4, pic5]
```

```
##cosine similarity between two lists
import numpy as np
from numpy import dot
from numpy.linalg import norm

def pair_cos(a,b):
  cos_sim = dot(a, b)/(norm(a)*norm(b))
  return cos_sim
```

```
##evaluation/ranking
from collections import defaultdict
rankings= defaultdict(list)
for i in range(5):
  for j in range(5):
    if i!=j:
      sim_score = pair_cos(np.array(pics[i]),np.array(pics[j]))
      rankings[i].append((j, sim_score))
  rankings[i].sort(key=lambda x:x[1], reverse=True)
```

```
images = []
for i in range(1, 6):
  imageread = plt.imread(f"{i}.png", format = "PNG")
  imageread = imageread[:100, :100, :]
  images.append(imageread)
  print('image shape: ', imageread.shape)
  plt.figure()
  plt.imshow(imageread)
  plt.show()
```

```
##evaluation/ranking
from collections import defaultdict
rankings= defaultdict(list)
for i in range(5):
  for j in range(5):
    if i!=j:
      sim_score = pair_cos(np.array(images[i].reshape(-1, 1)).T,np.array(images[j].reshape(-1, 1)))
      rankings[i].append((j, sim_score))
  rankings[i].sort(key=lambda x:x[1], reverse=True)
```
